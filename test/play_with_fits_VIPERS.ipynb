```python
from pyACC.fits import FitsManager
import numpy as np
import matplotlib.pyplot as plt
```

```python

imput_file = '/Users/matteocicirello/Desktop/astrocosmolib/test/VIPERS.fits' 
Vipers = FitsManager(imput_file)
Vipers.input_file
Vipers.hdulist.info()
data = Vipers.hdulist[1].data
Vipers.hdulist[1].header
print(data.columns)



```

```python

# Estrai i dati con i nomi corretti
zsp = data['zspec']
plt.plot(zsp, 'k.')
plt.xlabel('Galaxy ID')
plt.ylabel('Redshift')
plt.title('Redshifts of galaxies in the VIPERS survey')
plt.show()


```

```python
selmag = data['selmag']
plt.plot(selmag, 'k.')
plt.xlabel('Galaxy ID')
plt.ylabel('Magnitude')
plt.title('Magnitude of galaxies in the VIPERS survey')
plt.show()
```

```python
plt.plot(zsp, selmag, 'k.')
plt.plot(zsp, selmag, 'r.')
plt.xlabel('Redshift')
plt.ylabel('Magnitude')
plt.title('Redshift vs magnitude of galaxies in the VIPERS survey')
plt.show()
```

```python
plt.hist(zsp, bins=100, color='k')
plt.xlabel('Redshift')
plt.ylabel('Number of galaxies')
plt.title('Redshift distribution of galaxies in the VIPERS survey')
```
