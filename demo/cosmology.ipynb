```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append('../')
import matplotlib.pyplot as plt
import numpy as np
import astropy.units as u
from astropy.cosmology import LambdaCDM
from astropy.cosmology import w0waCDM
```

```python
from pyACC.cosmology import Distances
from pyACC.cosmology import HubbleFunction
```

```python
hubble_function = HubbleFunction( H0=70, omega_m=0.3, omega_l=0.7, omega_rad=0.0, omega_k=0.0)
```

```python
z = np.linspace(0, 1000, 1000)
```

```python
hubble_function(z)
```

```python
plt.plot(z, hubble_function(z),'r')
plt.xlabel('z')
plt.ylabel('H(z) [km/s/Mpc]')
plt.show()
```

```python
distances = Distances(hubble_function)
comoving_dist = distances.comoving_distance(z)
```

```python
plt.plot(z, comoving_dist)
plt.xlabel('z') 
plt.ylabel('Comoving distance [Mpc]')
plt.show()
```

```python
luminosity_dist = distances.luminosity_distance(z)
plt.plot(z, luminosity_dist)
plt.xlabel('z')
plt.ylabel('Luminosity distance [Mpc]')
plt.show()
```

```python
angular_dist = distances.angular_diameter_distance(z)
plt.plot(z, angular_dist)
plt.xlabel('z')
plt.ylabel('angular distance')
plt.show()  
```

```python
COSMOW0WACDM.H(z)
plt.plot(z, COSMOW0WACDM.H(z))
plt.xlabel('z')
plt.ylabel('H(z) [km/s/Mpc]')
plt.show()
```

```python
COSMOW0WACDM.angular_diameter_distance(z)
plt.plot(z, COSMOW0WACDM.angular_diameter_distance(z))
plt.xlabel('z')
plt.ylabel('angular distance')
plt.show()
```

```python
angular_distlcdm = cosmoLCDM.angular_diameter_distance(z)
angular_distw0wacdm = COSMOW0WACDM.angular_diameter_distance(z)
plt.figure(figsize=(8, 6))
plt.plot(z, angular_distlcdm, label="LambdaCDM")
plt.plot(z, angular_distw0wacdm, label="w0waCDM")
plt.xlabel("Redshift (z)")
plt.ylabel("Distanza angolare (Mpc)")
plt.title("Confronto tra modelli cosmologici")
plt.legend()
plt.show()
```

```python
comoving_dist_distlcdm = cosmoLCDM.comoving_distance(z)
comoving_dist_distw0wacdm = COSMOW0WACDM.comoving_distance(z)
plt.figure(figsize=(8, 6))
plt.plot(z, comoving_dist_distlcdm, label="LambdaCDM")
plt.plot(z, comoving_dist_distw0wacdm, label="w0waCDM")
plt.plot(z, comoving_dist, label="HubbleFunction")
plt.xlabel("Redshift (z)")
plt.ylabel("Comovente (Mpc)")
plt.title("Confronto tra modelli cosmologici")
plt.legend()
plt.show()
```

```python
comoving_dist = comoving_dist * u.Mpc
differenza  = (-comoving_dist_distlcdm + comoving_dist)
plt.plot(z, differenza)
plt.xlabel('z')
plt.ylabel('Differenza [Mpc]')
plt.show()
```
