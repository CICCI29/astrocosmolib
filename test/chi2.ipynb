```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
omega_m_values , chi2_values = np.load("/Users/matteocicirello/Desktop/astrocosmolib/test/chi2_values.npy")
```

```python
plt.plot(omega_m_values, chi2_values)
plt.xlabel(r"$\Omega_m$")
plt.ylabel(r"$\chi^2$")
plt.show()
```

```python
log_l = np.exp(-0.5 * chi2_values)
plt.plot(omega_m_values, log_l)
plt.xlabel(r"$\Omega_m$")   
plt.ylabel(r"$\log L$")
plt.show()
```

```python
#metodo cumulativa
cumfunc = np.cumsum(log_l)
cumfunc = cumfunc / cumfunc[-1]
plt.plot(omega_m_values, cumfunc)
plt.xlabel(r"$\Omega_m$")
plt.ylabel(r"$\int L d\Omega_m$")
plt.show()
```
